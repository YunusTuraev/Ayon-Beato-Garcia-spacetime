```python
import numpy as np
import sympy as sp 
sp.init_printing()
import matplotlib.pyplot as plt
```

```python
a,q,r,θ,ϕ,M,Q,L = sp.symbols('a,q,r,θ,ϕ,M,Q,L')
```

```python
Σ = r**2 + a**2*sp.cos(θ)
Σ
```

```python
f = (1-(2*M*r*sp.sqrt(Σ))/((Σ+Q**2)**(3/2))+(Σ*Q**2)/((Σ+Q**2)**2))
f
```

```python
Eff = f*(1+L**2/r**2)
Eff
```

```python
# Q=0, θ=π/3 va a ning turli qiymatlari uchun

Eff1 = (((((Eff.subs(M,1)).subs(Q,0)).subs(θ,sp.pi/3))).subs(L,5)).subs(a,0)
Eff2 = (((((Eff.subs(M,1)).subs(Q,0)).subs(θ,sp.pi/3))).subs(L,5)).subs(a,0.2)
Eff3 = (((((Eff.subs(M,1)).subs(Q,0)).subs(θ,sp.pi/3))).subs(L,5)).subs(a,0.5)
Eff4 = (((((Eff.subs(M,1)).subs(Q,0)).subs(θ,sp.pi/3))).subs(L,5)).subs(a,0.97)

print(Eff1)
print(Eff2)
print(Eff3)
print(Eff4)
```

```python
r = np.linspace(0.2,6,1000)

Eff1 = (1 + 25/r**2)*(-2*r/(r**2)**1.0 + 1)
Eff2 = (1 + 25/r**2)*(-2*r/(r**2 + 0.02)**1.0 + 1)
Eff3 = (1 + 25/r**2)*(-2*r/(r**2 + 0.125)**1.0 + 1)
Eff4 = (1 + 25/r**2)*(-2*r/(r**2 + 0.47045)**1.0 + 1)



#plt.figure(figsize=(8,5), dpi=200)

ax = plt.axes()


plt.xlabel("r/M")
plt.ylabel("$E_{eff}$")
plt.grid(True)

plt.xlim(0,5)
plt.ylim(-10,5)

plt.xticks(np.arange(0, 6, step=1))

plt.plot(r,Eff1, '--', label="Q=0, L=5", color='black', lw='1.2' )
plt.plot(r,Eff2, '--', label="Q=0.2, L=5", color='red', lw='1.2')
plt.plot(r,Eff3, '--', label="Q=0.5, L=5", color='green', lw='1.2')
plt.plot(r,Eff4, '--', label="Q=0.97, L=5", color='blue', lw='1.2' )

plt.legend(loc="upper right")

plt.show()
```
