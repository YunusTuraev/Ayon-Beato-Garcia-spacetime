```python
# Eff1 dagi ISCO radius
from scipy.optimize import brentq
import matplotlib.pyplot as plt

def f(r):
    return -(4.0*(1 + 25/r**2)/(r**2)**1.5 + 200.0/(r**2*(r**2)**1.5) + 150*(2*r**2/(r**2)**1.5 - 1)/r**4)

r = np.linspace(3, 10, 100)
plt.ylim(-0.1,0.1)
plt.plot(r, f(r))
plt.axhline(color = 'k')
root, info = brentq(f, 4, 5, full_output=True)
plt.axvline(x=root, color = 'r', **{'linestyle': 'dashed'})
print(root)
```
