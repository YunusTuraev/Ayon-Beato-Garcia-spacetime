```python
import matplotlib.pyplot as plt
import numpy as np 
import sympy as sp
```

```python
q,r = sp.symbols('q,r')
```

```python
z = r**4 + 3*q**2*r**2 + q**4 - 2*r**2*sp.sqrt(r**2+q**2)
z
```

```python
q = np.linspace(0,0.65)
r = np.linspace(0,2.1)
# z = np.linspace(0,0)
r,q = np.meshgrid(r,q)

z = r**4 + 3*q**2*r**2 + q**4 - 2*r**2*(r**2+q**2)**0.5

plt.figure(figsize=(8,5), dpi=200)
plt.grid(True)
plt.xlabel("Q/M")
plt.ylabel("r/M")
plt.title("R&Q parametric plot")

plt.xticks(np.arange(0, 1, step=0.5))
plt.yticks(np.arange(0, 2, step=0.5))

plt.contourf(q, r, z, 0)
plt.colorbar();
plt.show()
```
