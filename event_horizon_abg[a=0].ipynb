```python
import numpy as np
import sympy as sp
sp.init_printing()
import matplotlib.pyplot as plt
import scipy as sc
```

```python
r,θ,ϕ,M,Q,L = sp.symbols('r,θ,ϕ,M,Q,L') 
```

```python
f = 1-2*M*r**2/((r**2+Q**2)**1.5)+r**2*Q**2/((r**2+Q**2)**2)
f
```

```python
f1 = (f.subs(M,1)).subs(Q,0)
f2 = (f.subs(M,1)).subs(Q,0.5)
print(f1)
print(f2)
```

```python
e1 = sp.Eq(f1,0)
s1 = sp.solve(e1,r)
s1
```

```python
e2 = sp.Eq(f2,0)
s2 = sp.solve(e2,r)
s2
```

```python
from scipy.optimize import brentq

r=np.linspace(0,5)

f1 = -2*r**2/(r**2)**1.5 + 1
f2 = -2*r**2/(r**2 + 0.25)**1.5 + 0.25*r**2/(r**2 + 0.25)**2 + 1

plt.plot(r,f1, '--', color='green', lw=1.2)
plt.plot(r,f2, '--', color='blue', lw=1.2)

plt.xlim(0,3)
plt.ylim(-1,1)

plt.axhline(color = 'red')
#plt.axvline(r=root, color = 'r', **{'linestyle': 'dashed'})

plt.grid(True)

plt.show()
```

```python
# agar yuqoridagi (5) va (6) ishlamasa, quyidagi usulni qo'llang
# bu yerda f(1) uchun

from scipy.optimize import brentq
import matplotlib.pyplot as plt

def f(r):
    return -2*r**2/(r**2)**1.5 + 1
r = np.linspace(0, 3, 100)
plt.ylim(-1,1)
plt.plot(r, f(r))
plt.axhline(color = 'k')
root, info = brentq(f, 1.5, 2.5, full_output=True)
plt.axvline(x=root, color = 'r', **{'linestyle': 'dashed'})
print(root)
```
